ไม่อยากให้มี Header ก็เอาออกได้เลย

```
df5 = pd.read_csv('NFLX.csv', header = None)
```

```
df5.head()
```

หรือต้องการเปลี่ยน Header ให้อยู่รูปแบบอื่นๆ ก็แก้ไขได้

```
col_names = ['วัน', 'เปิด', 'สูงสุด', 'ต่ำสุด', 'ปิด', 'Adj Close', 'ปริมาณ']
df6 = pd.read_csv('NFLX.csv', header=None, names=col_names)
```

```
df6.head()
```
